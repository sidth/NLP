Importing modules

Coach contains the core training loop

MancalaGame contains the rules of the game and functions whose outputs we feed our neural network

NNet contains the neural network

```python
from Coach import Coach
from mancala.MancalaGame import MancalaGame as Game
from mancala.keras.NNet import NNetWrapper as nn
from utils import *
```

Arguments specifying 

1) Number of iterations

2) Number of episodes

3) Temperature threshold (we have only used temp threshold values of 0 and 1 in our algorithm

4) The fraction by which the trained network must beat the previous network for it to be accepted

5) Maximum lenght of the queue

6) Number of MCTS simulations per move

7) The number of games in which the new trained network is pit against the old one

8) Constant controlling the exploration in MCTS


9) Folder in which we save our weights

10) If we have pretrained models then put True o/w False

11) Location of weight file if we are starting midway

12) After these many iterations all the previous training examples are flushed out

```python
args = dotdict({
    'numIters': 5,
    'numEps': 4,
    'tempThreshold': 15,
    'updateThreshold': 0.6,
    'maxlenOfQueue': 200000,
    'numMCTSSims': 25,
    'arenaCompare': 20,
    'cpuct': 1,

    'checkpoint': './temp/',
    'load_model': False,
    'load_folder_file': ('/dev/models/8x100x50','best.pth.tar'),
    'numItersForTrainExamplesHistory': 20,

})
```

The main function

```python

if __name__=="__main__":
    g = Game(6)
    nnet = nn(g)

    if args.load_model:
        nnet.load_checkpoint(args.load_folder_file[0], args.load_folder_file[1])

    c = Coach(g, nnet, args)
    if args.load_model:
        print("Load trainExamples from file")
        c.loadTrainExamples()
    c.learn()
```
